```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
student_analysis = 'Copy of Newer Analysis - Student Summary.csv'
stud_summary = pd.read_csv(student_analysis)
```

```python
pd.options.display.max_columns = 1000
stud_summary.sort_values(by=['Total no. of sessions']);
```

```python
stud_summary = stud_summary[stud_summary['Result'] != 'Exempt']
```

```python
stud_summary = stud_summary.reset_index(drop=True);
```

```python
progress_log = pd.read_csv('Copy of Progress Log - Progress List.csv')
```

```python
progress_log.head()
```

```python
progress_log.sort_values(by=['Channel ID'], ascending = True, inplace = True)
progress_log = progress_log.reset_index(drop=True);
```

```python
progress_log["Progress Percentage"] = (progress_log['Chunk Number']/650 *100).round(2)
```

```python
stud_summary.rename(columns={'Total no. of sessions':'Sessions completed'}, inplace=True);
stud_summary.rename(columns={'Remaining sessions':'Sessions remaining'}, inplace=True);
```

```python
stud_full = pd.merge(stud_summary, progress_log[['Channel ID','Progress Percentage']], on=['Channel ID'])
```

```python
stud_full.columns.get_loc('< 330 avg chunks per session')
```

```python
stud_full.drop(stud_full.iloc[:, 13:25], axis=1, inplace=True)
stud_full.head()
```

```python
sessions = pd.read_csv('Student sessions as at 31 July 2018 (1) - Sessions.csv')
```

```python
sessions.dtypes
```

```python
sessions['Start time'] = pd.to_datetime(sessions['Start time']).apply(lambda x: x.time())
sessions['End time'] = pd.to_datetime(sessions['End time']).apply(lambda x: x.time())
```

```python
weekly_avg = pd.read_csv('Weekly sessions R - average sessions p week.csv')
```

```python
weekly_avg.drop([0], inplace = True);
```

```python
weekly_avg.rename(columns={'Unnamed: 0': 'Channel ID'}, inplace=True);
```

```python
weekly_avg.rename(columns={'avg ': 'weekly avg'}, inplace=True)
weekly_avg.head()
```

```python
weeks = []
df = stud_full

for row in df['GEMP']: 
    if row ==1.0:
        weeks.append(11)
    elif row==2.0:
        weeks.append(9)
    else:
        weeks.append('error')
        
df['Weeks Remaining'] = weeks
```

```python
df['Current chunk'] = df['Current chunk'].astype('int')
```

```python
df['Chunks remaining']= 660 - df['Current chunk']
```

```python
df = pd.merge(stud_full, weekly_avg, on=['Channel ID'])
```

```python
df['weekly avg'] = df['weekly avg'].round(2)
```

```python
df.head()
```

```python
df['Projected sessions'] = df['weekly avg'] * df['Weeks Remaining']
```

```python
df['Projected chunks'] = df['Avg chunks per session'] *df['Projected sessions']
```

```python
df['Session deficit'] = df['Sessions remaining'] - df['Projected sessions']
df['Chunk deficit'] = df['Chunks remaining'] - df['Projected chunks']
```

```python
df['Required chunks per week'] = df['Chunk deficit']/df['Weeks Remaining']
```

```python
df['Required sessions at chunk rate']= df['Chunk deficit']/df['Avg chunks per session']
```

```python
df['Required sessions per week']=df['Required sessions at chunk rate']/df['Weeks Remaining']
```

```python
over_sessions = []
over_sessions = (df['Required sessions per week']* df['Weeks Remaining']) > df['Sessions remaining']
over_sessions.value_counts()
```

```python
df['Over session limit?'] = over_sessions
```

```python
req1 = []

for row in df['Required sessions per week']: 
    if row <0:
        req1.append(0)
    else: 
        req1.append(row)
req1;
req1 = [round(x) for x in req1]
```

```python
df['Required sessions per week'] = req1
```

```python
df['Session rate'] = df['Required sessions per week'] - df['weekly avg'].astype(int)
```

```python
req4 = []
for row in df['Session rate']: 
    if row > 0:
        req4.append(row)
    elif row < 0:
        req4.append('0')
    else:
        req4.append('0')
req4;
```

```python
df['Session rate'] = req4
```

```python
df.head()
```

```python
for_students = df[['Student Number', 'Last Name','First Name','Result','Progress Percentage',\
                  'weekly avg','Avg chunks per session','Session rate', 'Over session limit?']].copy()
```

```python
for_students.to_csv('student report card R.csv', index=False)
```
