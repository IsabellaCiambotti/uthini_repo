```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sessions_formatted - sessions_formatted.csv')
```

```python
df.head()
```

```python
df['Start time'] = pd.to_datetime(df['Start time'])
df['End time']= pd.to_datetime(df['End time'])
```

```python
df['time delta'] = df['Start time']-df['End time']
```

```python
df = df.replace({'chunks accomplished': {0: 1}}) 
```

```python
df['Start hour']= 0
```

```python
df['Start hour'] = pd.to_datetime(df['Start time'])
```

```python
df['Start hour'] = df['Start hour'].dt.hour
```

```python
df.head(40)
```

```python
df['Start hour']= df['Start hour'].astype(int)
```

```python
import datetime as dt
tod = []
```

```python
for row in df['Start hour']:
    if row >= 6 and row < 12:
        tod.append('morning')
    elif row >=12 and row < 16:
        tod.append('midday')
    elif row >= 16 and row < 19:
        tod.append('afternoon')
    elif row >= 19 and row < 22:
        tod.append('evening')
    elif row >= 22:
        tod.append('night')
    elif row < 6:
        tod.append('night')
    else:
        tod.append('null') 
```

```python
df['Start hour'] = tod
```

```python
df.shape
```

```python
df2 = df[(df['chunks accomplished'] <= 30)]
```

```python
df2.shape
```

```python
df2['Start time R'] = pd.to_datetime(df2['Start time formatted'])
df2.drop(['Start time formatted', 'End time formatted'], axis=1);
```

```python
df2['weekday'] = df2['Start time R'].dt.weekday_name
```

```python
df2.to_csv('sessions_datetime.csv')
```
