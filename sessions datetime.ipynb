```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Actual start time (some, not all sessions) log - Log 1.csv')
```

```python
df.head()
```

```python
#Converting start time and end time to a readable format

df['ActualStartTime'] = pd.to_datetime(df['ActualStartTime'])
df['EndTime']= pd.to_datetime(df['EndTime'])
```

```python
#The time delta column defines the amount of time that a session takes

df['time delta'] = df['EndTime']-df['ActualStartTime']
```

```python
#Chunks accomplished defines how many chunks were completed per session

df['chunks accomplished']=df['EndChunk']-df['StartingChunk']
```

```python
# df = df.replace({'chunks accomplished': {0: 1}}) 
```

```python
df['Start hour']= 0
```

```python
#Defining a new column called Start Hour. We used this column to check which time of day the session was done in.

df['Start hour'] = pd.to_datetime(df['ActualStartTime'])
```

```python
df['Start hour'] = df['Start hour'].dt.hour
```

```python
pd.set_option('display.max_columns', None)
df.head(40)
```

```python
df['Start hour']= df['Start hour'].astype(int)
```

```python
import datetime as dt
tod = []
```

```python
#Defining the parameters to classify sessions by time of day

for row in df['Start hour']:
    if row >= 6 and row < 12:
        tod.append('morning')
    elif row >=12 and row < 16:
        tod.append('midday')
    elif row >= 16 and row < 19:
        tod.append('afternoon')
    elif row >= 19 and row < 22:
        tod.append('evening')
    elif row >= 22:
        tod.append('night')
    elif row < 6:
        tod.append('night')
    else:
        tod.append('null') 
```

```python
df['Start hour'] = tod
```

```python
df.shape
```

# Creating a new dataframe
Creating a new dataframe to remove sessions that consisted of more than 30 chunks because we considered them to be mistakes in the data

```python
df2 = df[(df['chunks accomplished'] <= 30)]
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2['Start time R'] = pd.to_datetime(df2['ActualStartTime'])
df2.drop(['ActualStartTime', 'EndTime'], axis=1);
```

```python
#Marking which day of the week each session took place
df2['weekday'] = df2['Start time R'].dt.weekday_name
```

```python
#Converting time delta into seconds so that we could flag those sessions that were too long
df2['time delta s'] = df2['time delta'].astype('timedelta64[s]')
```

```python
#Average seconds spent per weekday
delta_group = df2.groupby(['weekday'], as_index=False)['time delta s'].mean()
delta_group.head(40)
```

```python
#Flagging those sessions that are longer than 25 minutes (1500 seconds), because we considered them to be too long to be a single session
(df2['time delta s'] >= 1500.0).value_counts()
```
